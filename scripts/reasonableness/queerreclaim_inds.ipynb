```python
from liahr.benchmarks import QueerReclaimLex
ds_dir = ""
ds = QueerReclaimLex(root_dir=ds_dir, splits=["train", "test"], type="in", discard_ambiguous=True)
```

```python
inds = ds.annotator2label_inds["aggregate"]
```

```python
len(inds["harm"])
```

```python
import random
random.seed(None)
test_harm_inds = random.sample(inds["harm"], len(inds["harm"]) // 2)
train_harm_inds = list(set(inds["harm"]) - set(test_harm_inds))

test_noharm_inds = random.sample(inds["no harm"], len(test_harm_inds))
train_noharm_inds = list(set(inds["no harm"]) - set(test_noharm_inds))

train_inds = sorted(train_harm_inds + train_noharm_inds)
test_inds = sorted(test_harm_inds + test_noharm_inds)

print(f"train: {len(train_inds)}, test: {len(test_inds)}")
```
